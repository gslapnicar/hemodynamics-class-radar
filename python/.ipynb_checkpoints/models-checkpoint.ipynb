## Initial simple fully-connected architectures

We defined several different models, with different number of parameters and other design choices. These are in the following cells.

```python
def create_contact_fully_connected_small(input_data, lr, dropout, n_classes, metrics=["categorical_accuracy"], output_bias=None, regularization=0.001):
    n_instances = input_data.shape[0]
    input_shape = input_data.shape[1]
    n_branches = input_data.shape[2]
    if output_bias is not None:
        output_bias = tf.keras.initializers.Constant(output_bias)
    
    # 1.) create branch for contact BP
    input_contact_bp = Input(shape=(input_shape,), name='input_contact_bp')
    dense_contact_bp_1 = Dense(32, activation='relu', activity_regularizer=regularizers.l1(regularization))(input_contact_bp)
    dropout_bp_1 = Dropout(dropout)(dense_contact_bp_1)
    dense_contact_bp_2 = Dense(16, activation='relu', activity_regularizer=regularizers.l1(regularization))(dropout_bp_1)
    dropout_bp_2 = Dropout(dropout)(dense_contact_bp_2)
    final_bp = Dense(8, activation='relu', activity_regularizer=regularizers.l1(regularization))(dropout_bp_2)
    
    # 2.) create branch for contact ECG1
    input_contact_ecg1 = Input(shape=(input_shape,), name='input_contact_ecg1')
    dense_contact_ecg1_1 = Dense(32, activation='relu', activity_regularizer=regularizers.l1(regularization))(input_contact_ecg1)
    dropout_ecg1_1 = Dropout(dropout)(dense_contact_ecg1_1)
    dense_contact_ecg1_2 = Dense(16, activation='relu', activity_regularizer=regularizers.l1(regularization))(dropout_ecg1_1)
    dropout_ecg1_2 = Dropout(dropout)(dense_contact_ecg1_2)
    final_ecg1 = Dense(8, activation='relu', activity_regularizer=regularizers.l1(regularization))(dropout_ecg1_2)
    
    # 3.) create branch for contact ECG2
    input_contact_ecg2 = Input(shape=(input_shape,), name='input_contact_ecg2')
    dense_contact_ecg2_1 = Dense(32, activation='relu', activity_regularizer=regularizers.l1(regularization))(input_contact_ecg2)
    dropout_ecg2_1 = Dropout(dropout)(dense_contact_ecg2_1)
    dense_contact_ecg2_2 = Dense(16, activation='relu', activity_regularizer=regularizers.l1(regularization))(dropout_ecg2_1)
    dropout_ecg2_2 = Dropout(dropout)(dense_contact_ecg2_2)
    final_ecg2 = Dense(16, activation='relu', activity_regularizer=regularizers.l1(regularization))(dropout_ecg2_2)
    
    # 4.) create branch for contact ICG
    input_contact_icg = Input(shape=(input_shape,), name='input_contact_icg')
    dense_contact_icg_1 = Dense(32, activation='relu', activity_regularizer=regularizers.l1(regularization))(input_contact_icg)
    dropout_icg_1 = Dropout(dropout)(dense_contact_icg_1)
    dense_contact_icg_2 = Dense(16, activation='relu', activity_regularizer=regularizers.l1(regularization))(dropout_icg_1)
    dropout_icg_2 = Dropout(dropout)(dense_contact_icg_2)
    final_icg = Dense(8, activation='relu', activity_regularizer=regularizers.l1(regularization))(dropout_icg_2)
    
    # 5.) create branch for contact respiration
    input_contact_resp = Input(shape=(input_shape,), name='input_contact_resp')
    dense_contact_resp_1 = Dense(32, activation='relu', activity_regularizer=regularizers.l1(regularization))(input_contact_resp)
    dropout_resp_1 = Dropout(dropout)(dense_contact_resp_1)
    dense_contact_resp_2 = Dense(16, activation='relu', activity_regularizer=regularizers.l1(regularization))(dropout_resp_1)
    dropout_resp_2 = Dropout(dropout)(dense_contact_resp_2)
    final_resp = Dense(8, activation='relu', activity_regularizer=regularizers.l1(regularization))(dropout_resp_2)
    
    # 6.) create branch for contact z0
    input_contact_z0 = Input(shape=(input_shape,), name='input_contact_z0')
    dense_contact_z0_1 = Dense(32, activation='relu', activity_regularizer=regularizers.l1(regularization))(input_contact_z0)
    dropout_z0_1 = Dropout(dropout)(dense_contact_z0_1)
    dense_contact_z0_2 = Dense(16, activation='relu', activity_regularizer=regularizers.l1(regularization))(dropout_z0_1)
    dropout_z0_2 = Dropout(dropout)(dense_contact_z0_2)
    final_z0 = Dense(8, activation='relu', activity_regularizer=regularizers.l1(regularization))(dropout_z0_2)
    
    # merge branches
    merged = concatenate([final_bp, final_ecg1, final_ecg2, final_icg, final_resp, final_z0], name='concat')
    
    # unify learning part
    dense_learning_1 = Dense(32, activation='relu', activity_regularizer=regularizers.l1(regularization))(merged)
    dropout_learning_1 = Dropout(dropout)(dense_learning_1)
    dense_learning_2 = Dense(16, activation='relu', activity_regularizer=regularizers.l1(regularization))(dropout_learning_1)
    output = Dense(n_classes, activation='softmax', name='output', bias_initializer=output_bias)(dense_learning_2)
    model = Model(inputs=[input_contact_bp, 
                          input_contact_ecg1, 
                          input_contact_ecg2, 
                          input_contact_icg, 
                          input_contact_resp, 
                          input_contact_z0], 
                  outputs=output)
    
    opt = optimizers.Adam(learning_rate=lr)
    model.compile(optimizer=opt, loss='categorical_crossentropy', metrics=metrics)
    model.summary()
    plot_model(model, to_file="sanity_model_plots/model_contact_fully_connected_small.png", show_layer_names=True, show_shapes=True)
    
    return model
```

```python
def create_contact_fully_connected(input_data, lr, dropout, n_classes, metrics=["categorical_accuracy"]):
    n_instances = input_data.shape[0]
    input_shape = input_data.shape[1]
    n_branches = input_data.shape[2]
    
    # 1.) create branch for contact BP
    input_contact_bp = Input(shape=(input_shape,), name='input_contact_bp')
    dense_contact_bp_1 = Dense(256, activation='relu')(input_contact_bp)
    dropout_bp_1 = Dropout(dropout)(dense_contact_bp_1)
    dense_contact_bp_2 = Dense(128, activation='relu')(dropout_bp_1)
    dropout_bp_2 = Dropout(dropout)(dense_contact_bp_2)
    dense_contact_bp_3 = Dense(64, activation='relu')(dropout_bp_2)
    dropout_bp_3 = Dropout(dropout)(dense_contact_bp_3)
    dense_contact_bp_4 = Dense(32, activation='relu')(dropout_bp_3)
    dropout_bp_4 = Dropout(dropout)(dense_contact_bp_4)
    final_bp = Dense(32, activation='relu')(dropout_bp_4)
    
    # 2.) create branch for contact ECG1
    input_contact_ecg1 = Input(shape=(input_shape,), name='input_contact_ecg1')
    dense_contact_ecg1_1 = Dense(256, activation='relu')(input_contact_ecg1)
    dropout_ecg1_1 = Dropout(dropout)(dense_contact_ecg1_1)
    dense_contact_ecg1_2 = Dense(128, activation='relu')(dropout_ecg1_1)
    dropout_ecg1_2 = Dropout(dropout)(dense_contact_ecg1_2)
    dense_contact_ecg1_3 = Dense(64, activation='relu')(dropout_ecg1_2)
    dropout_ecg1_3 = Dropout(dropout)(dense_contact_ecg1_3)
    dense_contact_ecg1_4 = Dense(32, activation='relu')(dropout_ecg1_3)
    dropout_ecg1_4 = Dropout(dropout)(dense_contact_ecg1_4)
    final_ecg1 = Dense(32, activation='relu')(dropout_ecg1_4)
    
    # 3.) create branch for contact ECG2
    input_contact_ecg2 = Input(shape=(input_shape,), name='input_contact_ecg2')
    dense_contact_ecg2_1 = Dense(256, activation='relu')(input_contact_ecg2)
    dropout_ecg2_1 = Dropout(dropout)(dense_contact_ecg2_1)
    dense_contact_ecg2_2 = Dense(128, activation='relu')(dropout_ecg2_1)
    dropout_ecg2_2 = Dropout(dropout)(dense_contact_ecg2_2)
    dense_contact_ecg2_3 = Dense(64, activation='relu')(dropout_ecg2_2)
    dropout_ecg2_3 = Dropout(dropout)(dense_contact_ecg2_3)
    dense_contact_ecg2_4 = Dense(32, activation='relu')(dropout_ecg2_3)
    dropout_ecg2_4 = Dropout(dropout)(dense_contact_ecg2_4)
    final_ecg2 = Dense(32, activation='relu')(dropout_ecg2_4)
    
    # 4.) create branch for contact ICG
    input_contact_icg = Input(shape=(input_shape,), name='input_contact_icg')
    dense_contact_icg_1 = Dense(256, activation='relu')(input_contact_icg)
    dropout_icg_1 = Dropout(dropout)(dense_contact_icg_1)
    dense_contact_icg_2 = Dense(128, activation='relu')(dropout_icg_1)
    dropout_icg_2 = Dropout(dropout)(dense_contact_icg_2)
    dense_contact_icg_3 = Dense(64, activation='relu')(dropout_icg_2)
    dropout_icg_3 = Dropout(dropout)(dense_contact_icg_3)
    dense_contact_icg_4 = Dense(32, activation='relu')(dropout_icg_3)
    dropout_icg_4 = Dropout(dropout)(dense_contact_icg_4)
    final_icg = Dense(32, activation='relu')(dropout_icg_4)
    
    # 5.) create branch for contact respiration
    input_contact_resp = Input(shape=(input_shape,), name='input_contact_resp')
    dense_contact_resp_1 = Dense(256, activation='relu')(input_contact_resp)
    dropout_resp_1 = Dropout(dropout)(dense_contact_resp_1)
    dense_contact_resp_2 = Dense(128, activation='relu')(dropout_resp_1)
    dropout_resp_2 = Dropout(dropout)(dense_contact_resp_2)
    dense_contact_resp_3 = Dense(64, activation='relu')(dropout_resp_2)
    dropout_resp_3 = Dropout(dropout)(dense_contact_resp_3)
    dense_contact_resp_4 = Dense(32, activation='relu')(dropout_resp_3)
    dropout_resp_4 = Dropout(dropout)(dense_contact_resp_4)
    final_resp = Dense(32, activation='relu')(dropout_resp_4)
    
    # 6.) create branch for contact z0
    input_contact_z0 = Input(shape=(input_shape,), name='input_contact_z0')
    dense_contact_z0_1 = Dense(256, activation='relu')(input_contact_z0)
    dropout_z0_1 = Dropout(dropout)(dense_contact_z0_1)
    dense_contact_z0_2 = Dense(128, activation='relu')(dropout_z0_1)
    dropout_z0_2 = Dropout(dropout)(dense_contact_z0_2)
    dense_contact_z0_3 = Dense(64, activation='relu')(dropout_z0_2)
    dropout_z0_3 = Dropout(dropout)(dense_contact_z0_3)
    dense_contact_z0_4 = Dense(32, activation='relu')(dropout_z0_3)
    dropout_z0_4 = Dropout(dropout)(dense_contact_z0_4)
    final_z0 = Dense(32, activation='relu')(dropout_z0_4)
    
    # merge branches
    merged = concatenate([final_bp, final_ecg1, final_ecg2, final_icg, final_resp, final_z0], name='concat')
    
    # unify learning part
    dense_learning_1 = Dense(128, activation='relu')(merged)
    dropout_learning_1 = Dropout(dropout)(dense_learning_1)
    dense_learning_2 = Dense(64, activation='relu')(dropout_learning_1)
    output = Dense(n_classes, activation='softmax', name='output')(dense_learning_2)
    model = Model(inputs=[input_contact_bp, 
                          input_contact_ecg1, 
                          input_contact_ecg2, 
                          input_contact_icg, 
                          input_contact_resp, 
                          input_contact_z0], 
                  outputs=output)
    
    opt = optimizers.Adam(learning_rate=lr)
    model.compile(optimizer=opt, loss='categorical_crossentropy', metrics=metrics)
    model.summary()
    plot_model(model, to_file="sanity_model_plots/model_contact_fully_connected_large.png", show_layer_names=True, show_shapes=True)
    
    return model
```

```python
def create_radar_fully_connected(input_data, lr, dropout, n_classes, metrics=["categorical_accuracy"]):
    n_instances = input_data.shape[0]
    input_shape = input_data.shape[1]
    n_branches = input_data.shape[2]
    
    # 1.) create branch for radar I
    input_radar_i = Input(shape=(input_shape,), name='input_radar_i')
    dense_radar_i_1 = Dense(256, activation='relu')(input_radar_i)
    dropout_radar_i_1 = Dropout(dropout)(dense_radar_i_1)
    dense_radar_i_2 = Dense(128, activation='relu')(dropout_radar_i_1)
    dropout_radar_i_2 = Dropout(dropout)(dense_radar_i_2)
    dense_radar_i_3 = Dense(64, activation='relu')(dropout_radar_i_2)
    dropout_radar_i_3 = Dropout(dropout)(dense_radar_i_3)
    dense_radar_i_4 = Dense(32, activation='relu')(dropout_radar_i_3)
    dropout_radar_i_4 = Dropout(dropout)(dense_radar_i_4)
    final_radar_i = Dense(32, activation='relu')(dropout_radar_i_4)
    
    # 2.) create branch for radar Q
    input_radar_q = Input(shape=(input_shape,), name='input_radar_q')
    dense_radar_q_1 = Dense(256, activation='relu')(input_radar_q)
    dropout_radar_q_1 = Dropout(dropout)(dense_radar_q_1)
    dense_radar_q_2 = Dense(128, activation='relu')(dropout_radar_q_1)
    dropout_radar_q_2 = Dropout(dropout)(dense_radar_q_2)
    dense_radar_q_3 = Dense(64, activation='relu')(dropout_radar_q_2)
    dropout_radar_q_3 = Dropout(dropout)(dense_radar_q_3)
    dense_radar_q_4 = Dense(32, activation='relu')(dropout_radar_q_3)
    dropout_radar_q_4 = Dropout(dropout)(dense_radar_q_4)
    final_radar_q = Dense(32, activation='relu')(dropout_radar_q_4)
    
    # 3.) create branch for radar dist
    input_radar_dist = Input(shape=(input_shape,), name='input_radar_dist')
    dense_radar_dist_1 = Dense(256, activation='relu')(input_radar_dist)
    dropout_radar_dist_1 = Dropout(dropout)(dense_radar_dist_1)
    dense_radar_dist_2 = Dense(128, activation='relu')(dropout_radar_dist_1)
    dropout_radar_dist_2 = Dropout(dropout)(dense_radar_dist_2)
    dense_radar_dist_3 = Dense(64, activation='relu')(dropout_radar_dist_2)
    dropout_radar_dist_3 = Dropout(dropout)(dense_radar_dist_3)
    dense_radar_dist_4 = Dense(32, activation='relu')(dropout_radar_dist_3)
    dropout_radar_dist_4 = Dropout(dropout)(dense_radar_dist_4)
    final_radar_dist = Dense(32, activation='relu')(dropout_radar_dist_4)
    
    # 4.) create branch for radar pulse
    input_radar_pulse = Input(shape=(input_shape,), name='input_radar_pulse')
    dense_radar_pulse_1 = Dense(256, activation='relu')(input_radar_pulse)
    dropout_radar_pulse_1 = Dropout(dropout)(dense_radar_pulse_1)
    dense_radar_pulse_2 = Dense(128, activation='relu')(dropout_radar_pulse_1)
    dropout_radar_pulse_2 = Dropout(dropout)(dense_radar_pulse_2)
    dense_radar_pulse_3 = Dense(64, activation='relu')(dropout_radar_pulse_2)
    dropout_radar_pulse_3 = Dropout(dropout)(dense_radar_pulse_3)
    dense_radar_pulse_4 = Dense(32, activation='relu')(dropout_radar_pulse_3)
    dropout_radar_pulse_4 = Dropout(dropout)(dense_radar_pulse_4)
    final_radar_pulse = Dense(32, activation='relu')(dropout_radar_pulse_4)
    
    # 5.) create branch for radar resp
    input_radar_resp = Input(shape=(input_shape,), name='input_radar_resp')
    dense_radar_resp_1 = Dense(256, activation='relu')(input_radar_resp)
    dropout_radar_resp_1 = Dropout(dropout)(dense_radar_resp_1)
    dense_radar_resp_2 = Dense(128, activation='relu')(dropout_radar_resp_1)
    dropout_radar_resp_2 = Dropout(dropout)(dense_radar_resp_2)
    dense_radar_resp_3 = Dense(64, activation='relu')(dropout_radar_resp_2)
    dropout_radar_resp_3 = Dropout(dropout)(dense_radar_resp_3)
    dense_radar_resp_4 = Dense(32, activation='relu')(dropout_radar_resp_3)
    dropout_radar_resp_4 = Dropout(dropout)(dense_radar_resp_4)
    final_radar_resp = Dense(32, activation='relu')(dropout_radar_resp_4)
    
    # 6.) create branch for radar heart sound
    input_radar_heartSound = Input(shape=(input_shape,), name='input_radar_heartSound')
    dense_radar_heartSound_1 = Dense(256, activation='relu')(input_radar_heartSound)
    dropout_radar_heartSound_1 = Dropout(dropout)(dense_radar_heartSound_1)
    dense_radar_heartSound_2 = Dense(128, activation='relu')(dropout_radar_heartSound_1)
    dropout_radar_heartSound_2 = Dropout(dropout)(dense_radar_heartSound_2)
    dense_radar_heartSound_3 = Dense(64, activation='relu')(dropout_radar_heartSound_2)
    dropout_radar_heartSound_3 = Dropout(dropout)(dense_radar_heartSound_3)
    dense_radar_heartSound_4 = Dense(32, activation='relu')(dropout_radar_heartSound_3)
    dropout_radar_heartSound_4 = Dropout(dropout)(dense_radar_heartSound_4)
    final_radar_heartSound = Dense(32, activation='relu')(dropout_radar_heartSound_4)
    
    # merge branches
    merged = concatenate([final_radar_i, final_radar_q, final_radar_dist, final_radar_pulse, final_radar_resp, final_radar_heartSound], name='concat')
    
    # unify learning part
    dense_learning_1 = Dense(128, activation='relu')(merged)
    dropout_learning_1 = Dropout(dropout)(dense_learning_1)
    dense_learning_2 = Dense(64, activation='relu')(dropout_learning_1)
    output = Dense(n_classes, activation='softmax', name='output')(dense_learning_2)
    model = Model(inputs=[input_radar_i, 
                          input_radar_q, 
                          input_radar_dist, 
                          input_radar_pulse, 
                          input_radar_resp, 
                          input_radar_heartSound], 
                  outputs=output)
    
    opt = optimizers.Adam(learning_rate=lr)
    model.compile(optimizer=opt, loss='categorical_crossentropy', metrics=metrics)
    
    model.summary()
    plot_model(model, to_file="sanity_model_plots/model_radar_fully_connected_large.png", show_layer_names=True, show_shapes=True)
    
    return model
```

## Modular branch fully-connected architecture

Due to the fact that branched models cannot use the Keras functional API in traditional way (model.add()), we create functions that define each branch individually based on given hyperparameters.

```python
def single_fully_conn_branch(input_shape, hyperparams):
    # Always create an input layer + 1 Dense layer + 1 Dropout for each branch
    my_input = Input(shape=(input_shape,))
    first_dense = Dense(hyperparams["first_neuron"], activation=hyperparams["activation"])(my_input)
    l = Dropout(hyperparams["dropout"])(first_dense)
    
    # Add hidden dense layers + dropout to each branch
    for hidden_layer in range(hyperparams["hidden_layers"]):
        l = Dense(units=hyperparams["first_neuron"]/2, # triangular shape, but be careful -> units must not fall under 2 
                        activation=hyperparams["activation"], 
                        kernel_initializer=hyperparams["kernel_initializer"], 
                        activity_regularizer=hyperparams["regularizer"])(l)
        l = Dropout(hyperparams["dropout"])(l)
    
    final_branch_layer = Dense(units=hyperparams["first_neuron"], 
                               activation=hyperparams["activation"], 
                               kernel_initializer=hyperparams["kernel_initializer"], 
                               activity_regularizer=hyperparams["regularizer"])(l)
    
    return my_input, final_branch_layer
```

```python
def contact_fully_connected(x, y, x_val, y_val, hyperparams):
    n_instances = x.shape[0]
    input_shape = x.shape[1]
    n_branches  = x.shape[2]
    n_classes   = 6
    
    # Create each branch for each input
    branch_inputs  = []
    branch_outputs = []
    for branch in range(n_branches):
        curr_input, curr_output = single_fully_conn_branch(input_shape, hyperparams)
        branch_inputs.append(curr_input)
        branch_outputs.append(curr_output)
        
    # Always merge the branches
    l = concatenate(branch_outputs)
    

    # Additional dense layers
    for hidden_layer in range(hyperparams["hidden_layers"]):
        l = Dense(units=hyperparams["first_neuron"]/2, # triangular shape, but be careful -> units must not fall under 2 
                        activation=hyperparams["activation"], 
                        kernel_initializer=hyperparams["kernel_initializer"], 
                        activity_regularizer=hyperparams["regularizer"])(l)
        l = Dropout(hyperparams["dropout"])(l)

    # Always have output
    final_output = Dense(units=n_classes, 
                   activation=hyperparams["last_activation"],
                   kernel_initializer=hyperparams["kernel_initializer"], 
                   activity_regularizer=hyperparams["regularizer"])(l)
    
    # Define model
    model = Model(inputs=branch_inputs, outputs=final_output)
    
    model.compile(optimizer=hyperparams["optimizer"], loss="categorical_crossentropy", metrics=[
                    tf.keras.metrics.CategoricalCrossentropy(name="loss"),
                    tf.keras.metrics.TruePositives(name='tp'),
                    tf.keras.metrics.FalsePositives(name='fp'),
                    tf.keras.metrics.TrueNegatives(name='tn'),
                    tf.keras.metrics.FalseNegatives(name='fn'), 
                    tf.keras.metrics.CategoricalAccuracy(name="categorical_accuracy"),
                    tf.keras.metrics.Precision(name='precision'),
                    tf.keras.metrics.Recall(name='recall'),
                    tf.keras.metrics.AUC(name='auc'),
                ])
    model.summary()
    plot_model(model, to_file="sanity_model_plots/"+str(int(time.time()))+".png", show_layer_names=True, show_shapes=True)
    
    history = model.fit(x=[np.squeeze(x_branch) for x_branch in np.split(x, x.shape[2], axis=2)], 
                        y=to_categorical(y, num_classes=n_classes), 
                        validation_data=([np.squeeze(x_branch) for x_branch in np.split(x_val, x_val.shape[2], axis=2)], to_categorical(y_val, num_classes=n_classes)),
                        batch_size=hyperparams["batch_size"],
                        epochs=hyperparams["epochs"],
                        #callbacks=[ta.Live()],
                        verbose=0)
    
    return history, model
```

## Spectrogram architectures

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D
from keras.layers import Activation, Dropout, Flatten, Dense

def create_2d_CNN_small(input_data, kernel_size=(3,3), n_classes=6):
    
    # 1.) create branch for contact BP
    input_bp = Input(shape=(input_data.shape[0],input_data.shape[1],1))
    cnn_contact_bp_1 = Conv2D(16, kernel_size, activation="relu")(input_bp)
    max_pool_contact_bp_1 = MaxPooling2D(pool_size=(2, 2))(cnn_contact_bp_1)
    cnn_contact_bp_2 = Conv2D(8, kernel_size, activation="relu")(max_pool_contact_bp_1)
    max_pool_contact_bp_2 = MaxPooling2D(pool_size=(2, 2))(cnn_contact_bp_2)
    flat_bp = Flatten()(max_pool_contact_bp_2)
    
    # 2.) create branch for contact ECG1
    input_ecg = Input(shape=(input_data.shape[0],input_data.shape[1],1))
    cnn_contact_ecg_1 = Conv2D(16, kernel_size, activation="relu")(input_ecg)
    max_pool_contact_ecg_1 = MaxPooling2D(pool_size=(2, 2))(cnn_contact_ecg_1)
    cnn_contact_ecg_2 = Conv2D(8, kernel_size, activation="relu")(max_pool_contact_ecg_1)
    max_pool_contact_ecg_2 = MaxPooling2D(pool_size=(2, 2))(cnn_contact_ecg_2)
    flat_ecg = Flatten()(max_pool_contact_ecg_2)
    
    # 3.) create branch for contact ECG2
    input_ecg2 = Input(shape=(input_data.shape[0],input_data.shape[1],1))
    cnn_contact_ecg2_1 = Conv2D(16, kernel_size, activation="relu")(input_ecg2)
    max_pool_contact_ecg2_1 = MaxPooling2D(pool_size=(2, 2))(cnn_contact_ecg2_1)
    cnn_contact_ecg2_2 = Conv2D(8, kernel_size, activation="relu")(max_pool_contact_ecg2_1)
    max_pool_contact_ecg2_2 = MaxPooling2D(pool_size=(2, 2))(cnn_contact_ecg2_2)
    flat_ecg2 = Flatten()(max_pool_contact_ecg2_2)
    
    # 4.) create branch for contact ICG
    input_icg = Input(shape=(input_data.shape[0],input_data.shape[1],1))
    cnn_contact_icg_1 = Conv2D(16, kernel_size, activation="relu")(input_icg)
    max_pool_contact_icg_1 = MaxPooling2D(pool_size=(2, 2))(cnn_contact_icg_1)
    cnn_contact_icg_2 = Conv2D(8, kernel_size, activation="relu")(max_pool_contact_icg_1)
    max_pool_contact_icg_2 = MaxPooling2D(pool_size=(2, 2))(cnn_contact_icg_2)
    flat_icg = Flatten()(max_pool_contact_icg_2)
    
    # 5.) create branch for contact respiration
    input_resp = Input(shape=(input_data.shape[0],input_data.shape[1],1))
    cnn_contact_resp_1 = Conv2D(16, kernel_size, activation="relu")(input_resp)
    max_pool_contact_resp_1 = MaxPooling2D(pool_size=(2, 2))(cnn_contact_resp_1)
    cnn_contact_resp_2 = Conv2D(8, kernel_size, activation="relu")(max_pool_contact_resp_1)
    max_pool_contact_resp_2 = MaxPooling2D(pool_size=(2, 2))(cnn_contact_resp_2)
    flat_resp = Flatten()(max_pool_contact_resp_2)
    
    # 6.) create branch for contact z0
    input_z0 = Input(shape=(input_data.shape[0],input_data.shape[1],1))
    cnn_contact_z0_1 = Conv2D(16, kernel_size, activation="relu")(input_z0)
    max_pool_contact_z0_1 = MaxPooling2D(pool_size=(2, 2))(cnn_contact_z0_1)
    cnn_contact_z0_2 = Conv2D(8, kernel_size, activation="relu")(max_pool_contact_z0_1)
    max_pool_contact_z0_2 = MaxPooling2D(pool_size=(2, 2))(cnn_contact_z0_2)
    flat_z0 = Flatten()(max_pool_contact_z0_2)
    
    # merge branches
    merged = concatenate([flat_bp, flat_ecg, flat_ecg2, flat_icg, flat_resp, flat_z0], name='concat')
    
    # unify learning part
    dense_learning_1 = Dense(16, activation='relu')(merged)
    dropout_learning_1 = Dropout(0.2)(dense_learning_1)
    dense_learning_2 = Dense(8, activation='relu')(dropout_learning_1)
    output = Dense(n_classes, activation='softmax', name='output')(dense_learning_2)
    model = Model(inputs=[input_bp, 
                          input_ecg, 
                          input_ecg2, 
                          input_icg, 
                          input_resp, 
                          input_z0], 
                  outputs=output)
    
    opt = optimizers.Adam(learning_rate=0.001)
    model.compile(optimizer=opt, loss='categorical_crossentropy', metrics=["categorical_accuracy"])
    model.summary()
    plot_model(model, to_file="sanity_model_plots/model_2d_cnn_small.png", show_layer_names=True, show_shapes=True)
    
    return model
```

```python
def single_1d_cnn_branch(input_shape, hyperparams):
    # Always the same TinySleepNet idea for each branch
    my_input = Input(shape=(input_shape,1,))
    
    # Representation learning (convolutional part)
    cnn_1 = Conv1D(filters=hyperparams["N_FILTERS"], kernel_size=hyperparams["Fs"]//2, strides=hyperparams["Fs"]//4, padding="same")(my_input)
    maxpool_1 = MaxPooling1D(pool_size=8, strides=8)(cnn_1)
    dropout_1 = Dropout(hyperparams["DROPOUT"])(maxpool_1)
    cnn_2 = Conv1D(filters=hyperparams["N_FILTERS"], kernel_size=hyperparams["KERNEL_SIZE"], strides=1, padding="same")(dropout_1)
    cnn_3 = Conv1D(filters=hyperparams["N_FILTERS"], kernel_size=hyperparams["KERNEL_SIZE"], strides=1, padding="same")(cnn_2)
    cnn_4 = Conv1D(filters=hyperparams["N_FILTERS"], kernel_size=hyperparams["KERNEL_SIZE"], strides=1, padding="same")(cnn_3)
    maxpool_2 = MaxPooling1D(pool_size=4, strides=4)(cnn_4)
    dropout_2 = Dropout(hyperparams["DROPOUT"])(maxpool_2)
    
    # Sequence learning (LSTM part)
    lstm_1 = LSTM(units=hyperparams["N_UNITS"])(dropout_2)
    final_branch_layer = Dropout(hyperparams["DROPOUT"])(lstm_1)
    
    #final_branch_layer = Dense(units=hyperparams["N_UNITS"], activation=hyperparams["ACTIVATION"])(dropout_2)
    
    return my_input, final_branch_layer
```
