## Data preparation functions

### Input signals (X)



In completely raw terms, we have radar data (raw I and Q components, In-Phase and Quadrature) and some contact sensors (two ECG channels, ICG, blood pressure, z0 impedance).

From the above, some other signals are also computed. Adding those to the completely raw signals, we have the following input signals:

1. **contact sensors and derivations**:
- *contact_bp*: obtained with a combination of 2 cuff-like devices on arms (refer to image above)
- *contact_ecg1*: obtained with electrodes on arms and legs (refer to image above)
- *contact_ecg2*: obtained with electrodes on chest (refer to image above)
- *contact_icg*: obtained with a band + electrodes in the area between chest and stomach (refer to image above)
- *contact_z0*: obtained with same electrodes as ICG
- *contact_respiration*: derived from z0 

2. **radar and derivations**:
- *radar_i*: output of the radar, In-Phase component
- *radar_q*: output of the radar, Quadrature component
- *radar_dist*: computed with an algorithm from the I and Q components (https://www.nature.com/articles/s41598-018-29984-5)
- *radar_resp*: basically just filtered version of the distance
- *radar_heartSound*: computed from radar distance with some proprietary algorithm
- *radar_pulse*: computed from radar distance with some proprietary algorithm (similar to heartSound, but with more obvious heart cycles)

Examples of signal windows are available in another [cell](#Sanity-checks).

### Labels (Y)

The raw labels are not given in terms of discrete values, but rather a continuous "electricity" signal is given, which contains button presses. These appear as drops of "electricity" signal to close to zero - in other words, extreme falling and rising edge (together) mark a single button press.

### Important:
Please note that the button presses are inconsistent in number, and thus also meaning! Here is a summary per scenario:

1. **Resting**: There is no change in the time series, so the whole signal can be labelled as "resting"
2. **Valsalva**: An even number of button presses is expected (typically 6). The maneouvre is being done between each 2 presses.
3. **Apnea**: Similar to Valsalva, an even number of button presses expected (typically 4) and breath is held between each 2 presses.
4. **TiltUp / TiltDown**: Table starts in horizontal / vertical position and is being raised / lowered.
- *0 button presses*: The recording of data started after the table reached its final position. Thus, the full data can be labelled as TiltUp / TiltDown.
- *1 button press*: The button pressed when the table reached its final position. Thus, the data after the button press can be labelled as TiltUp / TiltDown.
- *2 button presses*: The button was pressed when the table started tilting and when it reached its final position. Thus, the data from 2nd button press onward can be labelled as TiltUp / TiltDown.

```python
def remap_ground_truth(class_array_raw, class_idx, scenario, subject_id, class_map, plot=False):
    class_array_new = np.zeros(class_array_raw.shape)

    if len(class_idx) % 2 != 0:
        print("[WARNING!] Odd number of indices!")
    else:
        if scenario == "Resting":
            class_array_new[:] = class_map[scenario]
        elif scenario == "Valsalva" or scenario == "Apnea":
            for i1, i2 in zip(class_idx[::2], class_idx[1::2]):
                class_array_new[i1:i2] = class_map[scenario]
        elif scenario == "TiltUp" or scenario == "TiltDown":
            if class_idx.size == 0:
                class_array_new[:] = class_map[scenario]
            elif class_idx.size == 1:
                class_array_new[class_idx[0]:] = class_map[scenario]
            elif class_idx.size == 2:
                if class_idx[0] == 0 and class_idx[1] == len(class_array_raw):
                    class_array_new[:] = class_map[scenario]
                elif class_idx[0] != 0 and class_idx[1] == len(class_array_raw):
                    class_array_new[class_idx[0]:] = class_map[scenario]
                else:
                    class_array_new[class_idx[1]:] = class_map[scenario]

    if plot:
        if not os.path.exists("sanity_signal_plots/" + subject_id + "/" + scenario):
            os.makedirs("sanity_signal_plots/" + subject_id + "/" + scenario)
        
        #print(class_idx)
        fig = plt.figure(figsize=(16, 12))
        l1, = plt.plot(class_array_raw.tolist())
        l2 = plt.vlines(class_idx, 0, 5, colors='r')
        l3, = plt.plot(class_array_new.tolist(), 'k')
        plt.title(subject_id + " " + scenario + " - class value")
        plt.ylim([-1, 6])
        plt.legend([l1, l2, l3], ['Electrical signal', 'Button presses', 'New label'])
        plt.savefig("sanity_signal_plots/" + subject_id + "/" + scenario + "/class.png")
        plt.close(fig)
        
    return class_array_new
```

This function below reshapes the time series into a 2D matrix:
- **rows**: each row is one window or instance for ML, which is of selected *window length*
- **columns**: each column is a single sample of the signal

**NOTE**: Overlap is chosen, currently at 1/2 of the window length. 

```python
def get_instances_sliding_window(array, window_width, overlap):
    return skimage.util.view_as_windows(array, window_width, step=overlap)
```

```python
def butter_bandpass(lowcut, highcut, fs, order=1):
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(order, [low, high], btype='band')
    return b, a


def butter_bandpass_filter(data, lowcut, highcut, fs, order=1):
    b, a = butter_bandpass(lowcut, highcut, fs, order=order)
    y = lfilter(b, a, data)
    return y
```

```python
def get_band(signal_name):
    # Set filtering band based on type of signal
    if signal_name=="contact_bp":
        band = [0.1, 4.0] # bp
    elif signal_name=="contact_ecg1":
        band = [0.1, 20.0] # ecg1
    elif signal_name=="contact_ecg2":
        band = [0.1, 20.0] # ecg2
    elif signal_name=="contact_icg":
        band = [0.1, 10.0] # icg
    elif signal_name=="contact_respiration":
        band = [0.1, 1.0] # respiration
    elif signal_name=="contact_z0":
        band = [0.1, 1.0] # z0, related to respiration
    elif signal_name=="radar_dist":
        band = [0.1, 1.0] # distance
    elif signal_name=="radar_heartSound":
        band = [0.1, 10.0] # heart sound
    elif signal_name=="radar_i":
        band = [0.1, 5.0] # i
    elif signal_name=="radar_pulse":
        band = [0.1, 3.0] # pulse
    elif signal_name=="radar_q":
        band = [0.1, 5.0] # q
    elif signal_name=="radar_resp":
        band = [0.1, 1.0] # respiration
    
    return band
```

## Data preparation (X and Y)

The function below does the majority of work when going from many long 1D signals to getting short windows formatted in 3D arrays.

**Input**: It takes the prepared data from the MATLAB scripts (a single dict, where 1st level are subjects, 2nd level are scenarios and 3rd level are obtained signals as 1D time series).

**Functionality**: Based on the button presses it remaps the scenarios to discrete integer values. Then the windowing is done and the most common label in a window is selected as the singular label. Remapped label graphs are saved. Additionally, the spectrograms are computed and saved.

**Output**: The result are dicts (for contact and contact-free signals), where 1st level keys are subject IDs and values are 3D arrays (rows = instances / windows, columns = signal samples, 3rd dimension = different signals). Additionally, a dict of ground-truth scenario labels is also provided, where keys are subject IDs and values are 1D arrays (temporal label).

```python
def prepare_data(file_path, fs, win_width, NFFT, class_map, signal_name_map, normalise, plot):
    X_contact_3d, X_radar_3d, Y_3d, plot_labels_contact, plot_labels_radar = {}, {}, {}, {}, {}
    
    # read the original data
    f         = h5py.File(file_path, 'r')
    full_data = f.get('output_re')
    
    # iterate across subjects
    for (subject_id, scenarios) in full_data.items():
        print("=", subject_id)
        
        # iterate across scenarios for each subject
        scenarios_contact_sigs, scenarios_radar_sigs, scenarios_labels, scenarios_contact_signal_names, scenarios_radar_signal_names = [], [], [], [], []
        for scenario, signals in scenarios.items():
            print("===", scenario)

            # remap scenario "signal" into descrete class values based on button presses
            class_array_raw      = signals.get("class")[()].flatten()
            class_idx            = signals.get("class_idx")[()].flatten(order='F').astype(int)   
            class_array_discrete = remap_ground_truth(class_array_raw, class_idx, scenario, subject_id, class_map, plot=plot)
            
            # then cut it into sliding windows with half a window overlap
            class_matrix_windowed = get_instances_sliding_window(
                class_array_discrete, win_width*fs, math.floor((win_width*fs)/2)
            )

            scenarios_labels.append(pd.DataFrame(class_matrix_windowed).mode(axis=1).values[:, 0])

            # iterate across all signals for each scenario for each subject
            contact_signals, radar_signals, signal_contact_names, signal_radar_names = [], [], [], []
            for signal_name, data in signals.items():
                if signal_name not in ["class", "class_idx"]:
                    if normalise:
                        signal = (data[()].flatten() - np.min(data[()].flatten())) / np.ptp(data[()].flatten())
                        band = get_band(signal_name)
                        signal = butter_bandpass_filter(signal, band[0], band[1], fs)
                    else:
                        signal = data[()].flatten()
                        band = get_band(signal_name)
                        signal = butter_bandpass_filter(signal, band[0], band[1], fs)
                    
                    # cut it into sliding windows with half a window overlap
                    signal_matrix_windowed = get_instances_sliding_window(
                        signal, win_width*fs, math.floor((win_width*fs)/2)
                    )

                    if "contact" in signal_name:
                        contact_signals.append(signal_matrix_windowed)
                        signal_contact_names.append(np.full(signal_matrix_windowed.shape, signal_name_map[signal_name]))
                    elif "radar" in signal_name:
                        radar_signals.append(signal_matrix_windowed)
                        signal_radar_names.append(np.full(signal_matrix_windowed.shape, signal_name_map[signal_name]))
                    else:
                        print("Anomaly!?")

            scenarios_contact_sigs.append(np.stack(contact_signals, axis=2))
            scenarios_radar_sigs.append(np.stack(radar_signals, axis=2))
            scenarios_contact_signal_names.append(np.stack(signal_contact_names, axis=2))
            scenarios_radar_signal_names.append(np.stack(signal_radar_names, axis=2))

        # Save the concatenated arrays to corresponding subject IDs
        X_contact_3d[subject_id]        = np.concatenate(scenarios_contact_sigs)
        X_radar_3d[subject_id]          = np.concatenate(scenarios_radar_sigs)
        Y_3d[subject_id]                = np.concatenate(scenarios_labels)
        plot_labels_contact[subject_id] = np.concatenate(scenarios_contact_signal_names)
        plot_labels_radar[subject_id]   = np.concatenate(scenarios_radar_signal_names)
        print()
    
    return X_contact_3d, X_radar_3d, Y_3d, plot_labels_contact, plot_labels_radar 
```

Set some basic parameters for the preparation, set the data location and run the function. It will output which subjects are being traveresed and which scenarios of each subject exists in the original data.

**Result:** We expect to obtain 3 dictionaries, one for contact, second for contact-free and third for the ground-truth labels. Additionally, some extra dicts (for plotting purposes) are provided.

```python
FILE_PATH = "../matlab/existing_code_starting_point/outputs/output_full_data_re.mat"
PLOT      = True
NORMALISE = True
FS_RE     = 100 # everything in _re files downsampled to 100 Hz (done in matlab, using shape-preserving piecewise cubic interpolation)
WIN_WIDTH = 20  # in seconds
NFFT      = 512 # for spectrograms

CLASS_MAP = {
    "Other":    0,
    "Resting":  1,
    "Valsalva": 2,
    "Apnea":    3,
    "TiltUp":   4,
    "TiltDown": 5
}

SIGNAL_NAME_MAP = {
    "contact_bp":         0,
    "contact_ecg1":       1,
    "contact_ecg2":       2,
    "contact_icg":        3,
    "contact_respiration":4,
    "contact_z0":         5,
    "radar_dist":         6,
    "radar_heartSound":   7,
    "radar_i":            8,
    "radar_pulse":        9,
    "radar_q":            10,
    "radar_resp":         11
}

X_contact_3d, X_radar_3d, Y_3d, plot_labels_contact, plot_labels_radar = prepare_data(
    file_path      = FILE_PATH, 
    fs             = FS_RE, 
    win_width      = WIN_WIDTH, 
    NFFT           = NFFT,
    class_map      = CLASS_MAP,
    signal_name_map= SIGNAL_NAME_MAP,
    normalise      = NORMALISE, 
    plot           = PLOT
)
```

## Spectrogram analysis

We consider a spectrogram input into deep learning, so instead of having a 1D time series (signal) input, we now have a 2D time-frequency (spectrogram) inputs into each branch for each signal.

```python
def get_spectrogram_scipy(x, y, fs, window_type, window_len, overlap, nfft, spec_plot_l, plot=False):
    f, t, Sxx = spectrogram(x, fs=fs, window=window_type, nperseg=window_len, noverlap=overlap, nfft=nfft)
    
    if plot:
        print("Spectrogram shapes:", f.shape, t.shape, Sxx.shape, "X and Y shapes:", x.shape, y.shape)
        
        fig = plt.figure(figsize=(16,6))
        ax1 = fig.add_subplot(2,1,1)
        plt.plot(x)
        plt.xlim([0,x.shape[0]])
        plt.title("Original signal (and filtered)")
        ax2 = fig.add_subplot(2,1,2)
        plt.imshow(Sxx[0:spec_plot_l, :], aspect='auto', cmap='hot_r', origin='lower')
        plt.plot(y)
        plt.title("Spectrogram (scipy) and ground truth")
        plt.show()
    
    print(Sxx.shape)
    return f, t, Sxx
    
def get_stft_scipy(x, y, fs, window_type, window_len, overlap, nfft, spec_plot_l, plot=False):
    f, t, Zxx = np.abs(stft(x, fs=fs, window=window_type, nperseg=window_len, noverlap=overlap, nfft=nfft))
    
    if plot:
        print("Spectrogram shapes:", f.shape, t.shape, Zxx.shape, "X and Y shapes:", x.shape, y.shape)
        
        fig = plt.figure(figsize=(16,6))
        ax1 = fig.add_subplot(2,1,1)
        plt.plot(x)
        plt.xlim([0,x.shape[0]])
        plt.title("Original signal (and filtered)")
        ax2 = fig.add_subplot(2,1,2)
        plt.imshow(Zxx[0:spec_plot_l, :], aspect='auto', cmap='hot_r', origin='lower')
        plt.plot(y)
        plt.title("STFT (scipy) and ground truth")
        plt.show()
    
    print(Zxx.shape)
    return f, t, Zxx

def get_spectrogram_wang(x, y, fs, window_type, window_len, overlap, nfft, spec_plot_l, plot=False):
    
    L = window_len # window size
    s = np.zeros((x.shape[0] - L + 1, L))
    
    # we slide the window by 1 sample
    if overlap == window_len-1:
        for i in range(x.shape[0] - L + 1):
            p = x[i:i + L]
            if window_type == "hamming":
                s[i, :] = np.divide(np.multiply(windows.hamming(L), (np.subtract(p, np.mean(p)))), (sys.float_info.epsilon + np.std(p)))
            else:
                print("Other window types not implemented atm.")
                return None
    else:
        print("Other overlap options not implemented atm.")
    
    s = np.abs(fft.fft2(s, s=nfft, axes=1))
    spec = s[:, 0:int(nfft/2)].T
    
    if plot:
        fig = plt.figure(figsize=(16, 6))
        ax1 = fig.add_subplot(2,1,1)
        plt.plot(x)
        plt.xlim([0,x.shape[0]])
        plt.title("Original signal (and filtered)")
        ax2 = fig.add_subplot(2,1,2)
        plt.imshow(spec[0:spec_plot_l, :], aspect='auto', cmap='viridis', origin='lower')
        plt.plot(y)
        plt.title('Spectrogram (Wenjin method) and ground truth:', fontsize=14, loc='left')
        plt.show()
    
    return None, None, spec
```

```python
def get_3d_spec(Sxx_in, moments=None):
    
    if moments is not None:
        (base_mean, base_std, delta_mean, delta_std,
             delta2_mean, delta2_std) = moments
    else:
        base_mean, delta_mean, delta2_mean = (0, 0, 0)
        base_std, delta_std, delta2_std = (1, 1, 1)
    
    h, w = Sxx_in.shape
    right1 = np.concatenate([Sxx_in[:, 0].reshape((h, -1)), Sxx_in], axis=1)[:, :-1]
    delta = (Sxx_in - right1)[:, 1:]
    delta_pad = delta[:, 0].reshape((h, -1))
    delta = np.concatenate([delta_pad, delta], axis=1)
    
    right2 = np.concatenate([delta[:, 0].reshape((h, -1)), delta], axis=1)[:, :-1]
    delta2 = (delta - right2)[:, 1:]
    delta2_pad = delta2[:, 0].reshape((h, -1))
    delta2 = np.concatenate([delta2_pad, delta2], axis=1)
    
    base = (Sxx_in - base_mean) / base_std
    delta = (delta - delta_mean) / delta_std
    delta2 = (delta2 - delta2_mean) / delta2_std
    stacked = [arr.reshape((h, w, 1)) for arr in (base, delta, delta2)]
    
    return np.concatenate(stacked, axis=2)
```

```python
def get_moments(x, y, fs, spec_method, w_type, w_len, overlap, nfft, band, spec_plot_l, sample_size=10, plot=False):
    nrows = x.shape[0]
    start_list = [random.randint(0, nrows) for i in range(sample_size)]
    if spec_method == "scipy_spec":
        Sxx_samples = [get_spectrogram_scipy(x[start,:], y[start], fs, window_type=w_type, window_len=w_len, overlap=overlap, nfft=nfft, filt_band=band, spec_plot_l=spec_plot_l, plot=False)[2] for start in start_list]
    elif spec_method == "scipy_stft":
        Sxx_samples = [get_stft_scipy(x[start,:], y[start], fs, window_type=w_type, window_len=w_len, overlap=overlap, nfft=nfft, filt_band=band, spec_plot_l=spec_plot_l, plot=False)[2] for start in start_list]
    elif spec_method == "wang_spec":
        Sxx_samples = [get_spectrogram_wang(x[start,:], y[start], fs, window_type=w_type, window_len=w_len, overlap=overlap, nfft=nfft, filt_band=band, spec_plot_l=spec_plot_l, plot=False)[2] for start in start_list]
    
    sxx_h, sxx_w = Sxx_samples[0].shape
    Sxx_3d_samples = np.array([get_3d_spec(Sxx) for Sxx in Sxx_samples])
    
    base_mean = Sxx_3d_samples[:, :, :, 0].mean()
    base_std = Sxx_3d_samples[:, :, :, 0].std()
    
    delta_mean = Sxx_3d_samples[:, :, :, 1].mean()
    delta_std = Sxx_3d_samples[:, :, :, 1].std()
    
    delta2_mean = Sxx_3d_samples[:, :, :, 2].mean()
    delta2_std = Sxx_3d_samples[:, :, :, 2].std()
    
    return (sxx_h, sxx_w), (base_mean, base_std, delta_mean, delta_std, delta2_mean, delta2_std)
```
