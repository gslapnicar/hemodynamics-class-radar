```python
%matplotlib inline

import matplotlib.pylab as plt
import numpy as np
import pandas as pd

from tsfresh import extract_features, extract_relevant_features, select_features
from tsfresh.utilities.dataframe_functions import impute
from tsfresh.feature_extraction import ComprehensiveFCParameters, EfficientFCParameters, MinimalFCParameters

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
WHICH_DATA = "contact"

X_all = np.load("model_inputs_and_targets/kfold/X_all_" + WHICH_DATA + ".npy")
Y_all = np.load("model_inputs_and_targets/kfold/Y_all.npy")
print("Original shape:", X_all.shape, Y_all.shape)
```

```python
X_feats = np.zeros(shape=(X_all.shape[0], 773, X_all.shape[2]))

for signal in range(X_all.shape[2]):
    print("Started signal ", signal)
    signal_df = pd.DataFrame(data=X_all[:, :, signal])
    display(signal_df.head())
    for i, row in signal_df.iterrows():
        df = pd.DataFrame({"id": [signal]*len(row.values), "signal": row.values})
        X_tsfresh = extract_features(df, column_id="id", default_fc_parameters=EfficientFCParameters(), disable_progressbar=True, n_jobs=8)
        X_feats[i, :, signal] = X_tsfresh
    
    print("Completed signal ", signal)
    break
```
